# 2. Univariate Analysis

Karena terdapat banyak missing value maka sebelum melakukan Analysis kita perlu melakukan data cleaning.
Berdasarkan keterangan data sebelumnya tidak terjadi keanehan pada data serta rata-rata dan median dalam setiap kolom sangat mendekati satu sama lain. Oleh karena itu, kami memutuskan untuk mengisi semua nilai yang hilang dengan median, karena beberapa feature akan lebih bermakna jika memiliki nilai bulat, seperti Complain, CouponUsed, OrderCount, dan sebagainya (tidak mungkin ada "Saya menggunakan kupon 1,75 kali" atau "Pengguna mengeluh sebanyak 0,2 kali")

```python
# Mengisi missing values
for col in df.columns:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)

# cek missing value
df.isnull().sum()
```
